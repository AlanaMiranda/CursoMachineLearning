VISUALIZAÇÃO DE DADOS COM MATPLOTLIB

```
# Importando as bibliotecas
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```
# use o método plot() para definir os valores dos eixos x e y, respectivamente 
plt.plot([1, 3, 5], [2, 5, 8])
plt.show()
```

```
# outra forma de criar um plot é atribuindo valores para as variaveis x e y
x = [1,5,10]
y = [2,4,8]
```

```
# construindo o plot
plt.plot(x,y)
# definindo rótulos para cada eixo
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
# definindo um título 
plt.title("Primeiro Plot")
plt.show()
```

```
x2=[1,2,3]
y2=[-1,-2,-3]
```

```
# definindo um label e passando como parâmetro do plot
plt.plot(x2,y2, label='Linha')
plt.legend()
plt.show()
```

Gráficos de Barras

```
x=[2,4,6,8,10]
y=[1,5,7,2,3]
```

```
# a função bar() cria o gráfico de barras
plt.bar(x,y, label='Barra', color='g')
plt.legend()
plt.show()
```

```
# criando um gráfico de barras duplas
x2=[1,3,5,7,9]
y2=[6,3,6,8,1]
```

```
plt.bar(x,y, label='Mulher', color='b')
plt.bar(x2,y2, label='Homem', color='y')
plt.legend()
```

Scatterplot (gráfico de dispersão)

```
x=[1,2,3,4,5]
y=[3,6,8,4,7]
```

```
plt.scatter(x,y, label='Pontos', color='r', marker=',', s= 100)
plt.legend()
plt.show()
```

Stack Plot (gráfico de área)

```
diaS=['seg','ter','qua','qui','sex']

dormir = [7,8,6,11,7]
comer = [2,3,4,3,2]
trabalhar = [7,8,7,2,2]
estudar = [8,5,7,8,13]
```

```
plt.plot([],[], color='green', label='Dormir', linewidth=5) 
plt.plot([],[],color='blue', label='Comer', linewidth=5) 
plt.plot([],[],color='pink', label='Trabalhar', linewidth=5)
plt.plot([],[],color='yellow', label='Estudar', linewidth=5)
plt.stackplot(diaS, dormir,comer,trabalhar, estudar, colors=['green', 'blue', 'pink', 'yellow'])
plt.xlabel('Dia da Semana')
plt.ylabel('Hora')
plt.legend()
plt.show()
```

Pie Chart (gráfico de pizza)

```
fatias = [2,5,10,7]
frutas= ['morango', 'maçã', 'banana', 'laranja']
cores = ['pink', 'red', 'yellow', 'orange']
```

```
plt.pie(fatias, labels=frutas, colors=cores, shadow=True, explode=(0,0,0.1,0))
plt.show()
```

PYLAB (Pyplot + Numpy)

```
from pylab import *
```

```
x = linspace(0, 5, 10)
y = x**2

# criando o objeto fig
fig= plt.figure()

# definindo os eixos e quanto eles devem ocupar (left, bottom, width, height) na tela, variando entre 0 e 1.
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Título')
```

```
# gráfico com duas figuras
x = linspace(0, 5, 10)
y = x**2

fig= plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #eixos da figura principal
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #eixos da figura secundária

# definindo a figura principal
axes1.plot(x,y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figura principal')
# definindo a figura secundária
axes2.plot(x,y, 'b')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('Figura secundária')
```
